# REMOVENDO OUTLIERS

```python
dados.boxplot(['price'])
```

```python
dados[dados['price'] >= 500000]
```

```python
 valor = dados['price']
```

```python
Q1 = valor.quantile(.25)
Q3 = valor.quantile(.75)
IIQ = Q3 - Q1
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 1.5 * IIQ 
```

```python
selecao = (valor >= limite_inferior) & (valor <= limite_superior)
dados_new = dados[selecao]
```

```python
dados_new.boxplot(['price'])
```

```python
grafico1_antes_de_retirar_outliers = dados.hist(['price'])
grafico2_depois_de_retirar_outliers = dados_new.hist(['price'])
```

```python
plt.savefig('dados/grafico1_antes_de_retirar_outliers.png', format='png')
plt.savefig('dados/grafico2_depois_de_retirar_outliers.png', format='png')
```
